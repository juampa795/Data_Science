##Primera entrega Final - Proyecto de DS

El presente proyecto buscará determinar el grado en que los clientes de un banco cuya información se trabajará, están dispuestos a salir de la institución dejando de utilizar los productos ofrecidos y en muchos casos recurrir a la competencia. Este tipo de comportamiento de los clientes, afecta las ganancias de la empresa y la participación que ocupa en el share del sistema financiero. Por esto es que poder determinar en cierto grado una predicción de los clientes que están dispuestos a salir de la entidad, nos permitirá llevar adelante politicas y campañas de retención de clientes de manera oportuna y reduciendo a cantidades aceptables las salidas. Los bancos pueden usar estos modelos predictivos para determinar cuales son los clientes que tienen cierto riesgo de salir de la entidad aplicando sobre estos estrategias de retención. Todo esto tambien beneficia a la institución para mantener un buen desempeño en la variable clientes activos, lo cual es muestra importante del share que ocupa en el sistema financiero. En el presente Dataset podemos observar la selección de 10.000 clientes de la institución bajo analisis, con un detalle sobre ciertas características y comportamientos de los mismos, lo que podremos utilizar para determinar mediante un analisis predictivo el grado en que estos estrían dispuestos a dejar de usar los productos del banco, ya por diferentes situaciones detectadas como un mal servicio de atención a los clientes, costos de mantenimiento de cuentas y productos que resulten elevados, problemas relacionados con la seguridad que brinda el banco en su homebanking y aplicaciones moviles, falta de disponibilidad de ATM y sucursales para facilitar acceso a atención personalizada, ya que hoy en día la omnicanalidad es muy importante a la hora de elegir mantenerse en una entidad.

Algunas de las preguntas que intentaré responder pueden ser:

La cantidad de productos que posee un cliente es un indicador de su baja propensión a salir del Banco?
Podemos idententificar algún patrón en el genero y/o ubicación del cliente que nos permita determinar una mayor propensión a salir del Banco?
Existe algún patrón en función a las edades de nuestros clientes que nos permitan observar si estos son más propensos a dejar sus productos?

Definición de Objetivos: El ojetivo del presente trabajo será identificar la/ las variables que posean una mayor correlación que nos permita determinar la mayor cantidad de salidas de los clientes del banco, para poder aplicar campañas comerciales adecuadas para reducir la cantidad de salidas.

Contexto Comercial: Nuestra institución Bancaria (caso de estudio) ha experimentado un incremento en la salida de los clientes, es decir los clientes que dejan de operar con nosotros ya sea cerrando sus cuentas o reduciendo su operatividad de manera que son considerados clientes inactivos, por lo que no representan ningun tipo de ganancia para el banco. En el presente proyecto se intentará determinar cual es el factor que influye de mayor manera en las salidas de los clientes.

Problema Comercial: La salida de nuestros clientes genera una perdida no solo de dinero sino del market share en cuanto a clientes activos. Además el costo de adquirir nuevos clientes, es más alto que tratar de mantener los que tenemos. La cantidad de nuestros clientes define el éxito de nuestro negocio.

Contexto Analítico: Como todos los negocios, el éxito se mide en función a los beneficios y a obtener dinero, mientras más tiempo mantengamos a nuestros clientes haciendo negocios con nosotros. más ganancias obtendremos. Además prestando mayor atención a las variables de salida podremos mejorar la experiencia de los clientes que cumplen con las variables que se analizarán. Para poder lograr todo esto primero empezaremos con un Analisis Exploratorio de los Datos (EDA), luego mediante algoritmos de Machine Learning tratare de definir cual es el mejor algoritmo para el presnta dataset.

```python
#Se podría plantear la hipotesis nula en la correlación de la edad y la salida de los clientes, en este caso sería que la edad no es un factor 
#que influye en la salida. Para obtener evidencias si la Hipotesis Nula puede ser cierta
import scipy.stats as stats 

grupo1 = df[df['Exited'] == 0]['Age'] #Defino las dos agrupaciones antes usadas
grupo2 = df[df['Exited'] == 1]['Age']

# Defino el t-test
t, p = stats.ttest_ind(grupo1, grupo2)
print('t = ', t)
print('p-value = ', p)
```
